## Importing libraries & Load Data 

```python
import numpy as np 
import pandas as pd 

import os
```

```python
df = pd.read_csv("./exchange_rates.csv")
df = df.drop(df.index[0:5]).dropna()
df.head(10)
```

```python
df.columns
```

```python
df = df[df != 'ND']
df.dropna()



from datetime import datetime

df[df.columns[0]] = pd.to_datetime(df[df.columns[0]]) 
df[df.columns[1:len(df.columns)]] = df[df.columns[1:len(df.columns)]].astype(float)
```

```python
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt

x = df[df.columns[0]]
y = df[df.columns[4]]
fig = plt.figure(figsize=(14, 6))

plt.plot(x,y,'b')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('CHINA -- SPOT EXCHANGE RATE')

plt.show()
```
