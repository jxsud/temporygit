````python
# import the pandas library

import pandas as pd
````

````python
# Load different sheets from the same excel file
df_GOOGL = pd.read_excel('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/market_data.xls', sheet_name='GOOGL', index_col='Date', parse_dates=True)
df_IBM = pd.read_excel('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/market_data.xls', sheet_name='IBM', index_col='Date', parse_dates=True)
````

````python
# Create small dataframes for demonstration
cols = ['High', 'Low']

df1 = df_IBM[cols].loc['2017-Jan-01' : '2017-Jan-17'].sort_index()
df2 = df_GOOGL[cols].loc['2017-Jan-10' : '2017-Jan-24'].sort_index()

# show both dataframes
print("== IBM ==")
display(df1)
print("== GOOGLE ==")
display(df2)
````

## Outer Merge

We will do an outer merge with ```indicator=True```

````python
# merge how=outer on=Date indicator=True
df_merge = df1.merge(df2, on='Date', how='outer', indicator=True)

df_merge
````

Now the new `_merge` column shows us which rows were in each of the original dataframes.

We can use this to select those rows that were **not** in both


````python
# now filter where _merge contains the string 'both'
df_anti_join = df_merge[ df_merge['_merge'] != 'both' ].copy()

df_anti_join
````

now drop the _merge column

````python
df_anti_join.drop('_merge', inplace=True, axis=1)
df_anti_join
````
